```python
# Dependencies
import requests
import pandas as pd
from pprint import pprint
import csv
import re
import matplotlib.pyplot as plt
import numpy as np


# Import API Key - optional
#import api_keys

#URL Components
url = "http://www.omdbapi.com/?t="
key = "3fb5f5bd"

#TEST!!!
movieTitle = "The Dark Knight"
combinedURL = url + movieTitle + "&apikey=" + key
#print (combinedURL)
response = requests.get(combinedURL).json()
pprint(response)
```

```python
imdbRatings = []
BoxOfficeReturns = []

#Import Movie List - use these titles in api query
movies = pd.read_csv('input csv/newmovielist.csv', encoding='utf8')['Film Name']

movies = [re.split(r'\s{2,}', s.strip())[0].lower() for s in movies]
print(movies)

movieList = list(movies)
#print(movieList)

for i in range(len(movieList)):
    combinedURL = url + str(movieList[i]) + "&apikey=" + key
    print(combinedURL)
    response = requests.get(combinedURL).json()
    print("Movie Title: " + str(movieList[i][0]))
    try:
        print("imdbRating: " + response["imdbRating"])
        imdbRatings.append(float(response["imdbRating"]))
    except:
        print("Coult not get IMDB Rating")
        imdbRatings.append(np.nan)        
    
    try:
        print("BoxOffice: " + response["BoxOffice"])
        BoxOfficeReturns.append(str(response["BoxOffice"]))
    except:
        print("Coult not get Box Office Return")
        BoxOfficeReturns.append(np.nan)         
    print("\n")        

print(len(movieList))
print(len(imdbRatings))
print(len(BoxOfficeReturns))
```

```python
##### MOVIE RATINGS

pd.DataFrame(movieList)
movieListDB = pd.DataFrame(
    {'Title': movieList,
     'Movie Rating': imdbRatings,
     'Movie Revenue': BoxOfficeReturns
    })

#movieListDB.fillna(0.0) 
movieListDB["Movie Revenue"] = movieListDB["Movie Revenue"].replace("N/A", np.nan)
movieListDB["Movie Revenue"] = movieListDB['Movie Revenue'].replace('[\$,)]','', regex=True).astype(float)

movieListDB.head()
```

```python
#####BOOK RATINGS

with open('input csv/booklist_ratings.csv', encoding = "utf-8", errors = "ignore") as f:
    reader = csv.reader(f)
    bookList = list(reader)

#bookList = [re.split(r'\s{2,}', s.strip())[0].lower() for s in bookList]
#print(bookList)
#for i in bookList:
 #   print (i)
    
bookListDB = pd.DataFrame(bookList)

bookListDB = bookListDB.rename(columns={1 : "Book Title", 2 : "Book Rating", 3 : "Average Votes"})
bookListDB = bookListDB.drop([0], axis=1)
bookTitles = list(bookListDB["Book Title"])

bookTitles = [re.split(r'\s{2,}', s.strip())[0].lower() for s in bookTitles]

#re.group(r'(.*)(\(.*\))')

bookListDB["Title"] = bookTitles
bookListDB = bookListDB.drop(["Book Title"], axis=1)

bookListDB
```

```python
CombinedDF = pd.merge(movieListDB, bookListDB, on="Title", how="left")
CombinedDF
```
