```python
# This script is to load the adapted science fiction books and films into MongoDB

from pymongo import MongoClient
import pandas as pd
```

```python
# Import booksDB

booksDB = pd.read_csv("Transformed_data/bookListDB.csv", encoding='utf8')
booksDB.head()
```

```python
conn = "mongodb://localhost:27017"
client = MongoClient(conn)
db = client.adapted_scifi_films_db
collection = db.books
```

```python
booksDB_dict = booksDB.to_dict(orient='records')
collection.insert_many(booksDB_dict)
```

```python
# Verify insertion to Mongo was okay:
results = collection.find()
for result in results:
    print(result)
```

```python
# Import moviesDB

moviesDB = pd.read_csv("Transformed_data/movieListDB.csv", encoding='utf8', engine='python').iloc[:,1:]
moviesDB.head()
```

```python
# Dump to MongoDB:
collection = db.movies
moviesDB_dict = moviesDB.to_dict(orient='records')
collection.insert_many(moviesDB_dict)
```

```python
# Verify:
results = collection.find()
for result in results:
    print(result)
```
