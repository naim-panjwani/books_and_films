```python
# Import dependencies
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import json
import os
```

```python
# Create Empty Lists
movie_list1 = []
movie_list2=[]
```

```python
#HTML Parsing of Wiki page
url = "https://en.wikipedia.org/wiki/Category:Films_based_on_science_fiction_novels"

response = requests.get(url)
soup = bs(response.text, 'html.parser')
results=soup.find_all("div", class_="mw-category-group")
for result in results:
    titles = result.find_all("li")
   
    for title in titles:
        new= title.find("a").text
        movie_list1.append([str(new.strip())])
        
print(movie_list1)
```

```python
#HTML Parsing
url = "https://en.wikipedia.org/w/index.php?title=Category:Films_based_on_science_fiction_novels&pagefrom=Village+Of+The+Damned+%281960+Film%29%0AVillage+of+the+Damned+%281960+film%29#mw-pages"

response = requests.get(url)
soup = bs(response.text, 'html.parser')

next_results=soup.find_all("div", class_="mw-category-group")

for next_result in next_results:
    next_titles = next_result.find_all("li")
   
    for next_title in next_titles:
        new_list= next_title.find("a").text
        movie_list2.append([str(new_list.strip())])
print(movie_list2)


```

```python
#Creating Combined Movie list
movie_list= movie_list1+movie_list2


#Creating Dataframe of Movie List
movie_list_df= pd.DataFrame(movie_list)
movie_list_df
```

```python
#Renaming Columns
movie_list_df.columns=['Film Name']
movie_list_df.head()
```

```python
#Drop Duplicates
new_movie_list_df = movie_list_df.drop_duplicates
new_movie_list_df
```

```python
#Import Results in flat file
new_movie_list_df.to_csv('movielist.csv',index=False, header=True)
```
