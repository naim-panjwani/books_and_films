```python
imdbRatings = []
BoxOfficeReturns = []

#IMPORT MOVIE LIST FROM CSV
movieList = pd.read_csv('input csv/newmovielist.csv', encoding='utf8')['Film Name']
#FORMAT MOVIE LIST TO LOWER CASE AND STRIP
movieList = [re.split(r'\s{2,}', s.strip())[0].lower() for s in movieList]
#PRINT MOVIE LIST, CONFIRM FORMATTING
print(movieList)
print("\n\n")

#RUN ALL MOVIES THROUGH API, RETRIEVE RATINGS AND BOX OFFICE RETURNS
for i in range(len(movieList)):
    combinedURL = url + str(movieList[i]) + "&apikey=" + key
    print(combinedURL)
    response = requests.get(combinedURL).json()
    print("Movie Title: " + str(movieList[i]))
    try:
        print("imdbRating: " + response["imdbRating"])
        imdbRatings.append(float(response["imdbRating"]))
    except:
        print("Coult not get IMDB Rating")
        imdbRatings.append(np.nan)        
    
    try:
        print("BoxOffice: " + response["BoxOffice"])
        BoxOfficeReturns.append(str(response["BoxOffice"]))
    except:
        print("Coult not get Box Office Return")
        BoxOfficeReturns.append(np.nan)         
    print("\n")        

#SHOW THE NUMBER OF ENTRIES INTO EACH LIST, VERIFY DF CAN BE CREATED
print("Movie Title Entries: \t" + str(len(movieList)))
print("Movie Rating Entries: \t" + str(len(imdbRatings)))
print("Movie Boxoffice Entries:" + str(len(BoxOfficeReturns)))
```

```python
##### MOVIE RATINGS

#CREATE DATAFRAME FROM LISTS
pd.DataFrame(movieList)
movieListDB = pd.DataFrame(
    {'Title': movieList,
     'Movie Rating': imdbRatings,
     'Movie Revenue': BoxOfficeReturns
    })

#REPLACE N/A'S WITH NaN's
movieListDB["Movie Revenue"] = movieListDB["Movie Revenue"].replace("N/A", np.nan)

#CHECK ON TYPE, FOR FORMATTING
#type(CombinedDF['Movie Revenue'][0])
#FORMAT REVENUES FROM STRINGS TO FLOATS NUMBER
movieListDB["Movie Revenue"] = movieListDB['Movie Revenue'].replace('[\$,)]','', regex=True).astype(float)

#SHOW DB HEAD
movieListDB.head()
```

```python
#####BOOK RATINGS

#RETRIEVE CSV
with open('input csv/merged_list.csv', encoding = "utf-8", errors = "ignore") as f:
    reader = csv.reader(f)
    bookList = list(reader)
    
#CREATE DATAFRAME FROM BOOKLIST CSV
bookListDB = pd.DataFrame(bookList)
#RENAME COLUMNS
bookListDB = bookListDB.rename(columns={1: "Movie Title", 2 : "Book Title", 3 : "Book Rating", 4 : "Average Votes"})

#STRIP TITLES AND CONVERT TO LOWER CASE
bookTitles = list(bookListDB["Movie Title"])
bookTitles = [re.split(r'\s{2,}', s.strip())[0].lower() for s in bookTitles]
#ADD FORMATTED TITLES TO DF AS COLUMN TITLE
bookListDB["Title"] = bookTitles

#DROP UNNECESSARY COLUMN
bookListDB = bookListDB.drop([0, "Movie Title"], axis=1)

#SHOW DB HEAD
bookListDB.head(10)
```

```python
CombinedDF = pd.merge(movieListDB, bookListDB, on="Title", how="left")
CombinedDF.head(10)
```

```python
#GET ENTRY COUNTS FOR EACH COLUMN
MovieRatingsCounts = CombinedDF['Movie Rating'].count()
MovierevenueCounts = CombinedDF['Movie Revenue'].count()
BookRatingsCounts = CombinedDF['Book Rating'].count()
BookAverageVotesCounts = CombinedDF['Average Votes'].count()
#SHOW ENTRY COUNTS
print ("Movie Rating: " + str(MovieRatingsCounts))
print ("Book Rating: " + str(BookRatingsCounts))
print ("Movie Revenue: " + str(MovierevenueCounts))
print ("Book Average Votes: " + str(BookAverageVotesCounts))
```
