```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import json
import os
```

```python
file_to_load = "newmovielist.csv"

# Read Purchasing File and store into Pandas data frame
movie_data = pd.read_csv(file_to_load)
movie_data
```

```python
booklist=[]
for movie in movie_data['Film Name']:
    film_name = movie.replace(' ', '+')
    url = 'https://www.goodreads.com/search?utf8=%E2%9C%93&query=' + str(film_name)+'&search_type=books'
    response = requests.get(url)
    soup = bs(response.text, 'lxml')
    for result in soup:
        try:
            title=soup.find('span',itemprop='name').text
            rating = soup.find('span', class_='minirating').text
            booklist.append([str(movie),str(title.strip()), str(rating.strip())])
        except:
            pass
    
df_books = pd.DataFrame(booklist)
df_books
```

```python
new_books= df_books.drop_duplicates()
new_books.columns=['movie','books','ratings']
new_books=new_books.reset_index(drop=True)
```

```python
def strType(var):
    try:
        if int(var) == float(var):
            return 'int'
    except:
        try:
            float(var)
            return 'float'
        except:
            return 'str'
def getRating(sentence):
    words = sentence.split(' ')
    for word in words:
        if strType(word.replace(',','')) == 'float':
            return(float(word.replace(',','')))
def getNumReviews(sentence):
    words = sentence.split(' ')
    for word in words:
        if strType(word.replace(',','')) == 'int':
            return(int(word.replace(',','')))
```

```python
ratings=[]
for i in new_books['ratings']:
    ratings.append([str(getRating(i)), str(getNumReviews(i))])

df=pd.DataFrame(ratings,columns=['rating','average_votes'])
df


    
```

```python
result = pd.concat([new_books,df],axis=1,sort=False)
final=result.drop(['ratings'],axis=1)
final
```

```python
final.to_csv('merged_list.csv')
```
