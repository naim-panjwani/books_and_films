```python
def searchInKaggle(search_string):
    # PRE: a valid search string to search books in book_reviews 'title' column dataframe
    # POST: returns the subset of the dataframe with matching results
    booktitles = [s.lower() for s in book_reviews['title']]
    matching = [s for s in booktitles if search_string.lower() in s]
    results = book_reviews.loc[booktitles.isin(matching)]
    return(results)

searchInKaggle('Action Replay')
```
